```python
import pandas as pd

dataset = pd.read_csv("insurance_pre.csv")

from sklearn.preprocessing import LabelEncoder

dataset['sex'] = LabelEncoder().fit_transform(dataset['sex'])
dataset['smoker'] = LabelEncoder().fit_transform(dataset['smoker'])

independent = dataset[['age', 'sex', 'bmi', 'children', 'smoker']]
dependent = dataset[['charges']]

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor #importing the Linear regression library from sklearn 
# regressor = RandomForestRegressor(n_estimators=500,criterion='absolute_error',
#                     max_depth = None , min_samples_split=10, min_samples_leaf=5, bootstrap=True)

regressor = RandomForestRegressor(n_estimators=100,criterion='absolute_error',
                    max_depth = None , min_samples_split=10, min_samples_leaf=5, bootstrap=True)

regressor.fit(X_train, Y_train)

Y_train_pred = regressor.predict(X_train)

Y_pred = regressor.predict(X_test)

from sklearn.metrics import r2_score # for r2 score evaluvation importing the function from metrics library

r2_score_train = r2_score(Y_train, Y_train_pred) 
print("train r2",r2_score_train)

r_score= r2_score(Y_test,Y_pred) # comparing the predicted values and test(Actual) values to calculate R2Score 
print("test r2",r_score)
```

```python
import pickle

filename = "Insurance_premium_predictor_Random_Forest.sav"
pickle.dump(regressor,open(filename, 'wb'))

loaded_model = pickle.load(open("Insurance_premium_predictor_Random_Forest.sav", 'rb'))
result = loaded_model.predict([[45,0,27.9,0,1]])
print(result)
```
