```python
import pandas as pd

dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
from sklearn.preprocessing import LabelEncoder

dataset['sex'] = LabelEncoder().fit_transform(dataset['sex'])
dataset['smoker'] = LabelEncoder().fit_transform(dataset['smoker'])

dataset
```

```python
independent = dataset[['age', 'sex', 'bmi', 'children', 'smoker']]
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

```python
from sklearn.svm import SVR #importing the Linear regression library from sklearn
from sklearn.preprocessing import StandardScaler

# regressor = SVR(kernel="linear", C=0.1)
# regressor = SVR(kernel="linear", C=1)
# regressor = SVR(kernel="linear", C=10)
# regressor = SVR(kernel="linear", C=100)
# regressor = SVR(kernel="linear", C=1000)

#regressor = SVR(kernel="rbf", C=0.1)
#regressor = SVR(kernel="rbf", C=1)
#regressor = SVR(kernel="rbf", C=10)
#regressor = SVR(kernel="rbf", C=100)
#regressor = SVR(kernel="rbf", C=1000)

#regressor = SVR(kernel="poly", degree=1, C=0.1)
#regressor = SVR(kernel="poly", degree=2, C=0.1)
regressor = SVR(kernel="poly", degree=3, C=1000)

# Initialize scaler
scaler = StandardScaler()

# Fit on training data and transform
X_train_scaled = scaler.fit_transform(X_train)

# Transform test data using the same scaler
X_test_scaled = scaler.transform(X_test)

regressor.fit(X_train_scaled, Y_train)

Y_train_pred = regressor.predict(X_train_scaled)

from sklearn.metrics import r2_score # for r2 score evaluvation importing the function from metrics library

r_score_train = r2_score(Y_train,Y_train_pred) # comparing the predicted values and test(Actual) values to calculate R2Score 
print("training r2",r_score_train)

Y_pred = regressor.predict(X_test_scaled)

from sklearn.metrics import r2_score # for r2 score evaluvation importing the function from metrics library

r_score= r2_score(Y_test,Y_pred) # comparing the predicted values and test(Actual) values to calculate R2Score 
print("test r2",r_score)
```

```python
import pickle

# Save both model and scaler
pickle.dump((regressor, scaler), open("Insurance_premium_predictor_SVR.sav", "wb"))

# Load both
loaded_model, loaded_scaler = pickle.load(open("Insurance_premium_predictor_SVR.sav", "rb"))

# Scale new input before prediction
new_input = [[45,0,27.9,0,0]]
new_input_scaled = loaded_scaler.transform(new_input)

result = loaded_model.predict(new_input_scaled)
print(result)
```
