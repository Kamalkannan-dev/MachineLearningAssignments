```python
import pickle

mlr_model = pickle.load(open("Insurance_premium_predictor_MLR.sav","rb"))

print("Premium predicted by MLR model",mlr_model.predict([[40,0,27.9,0,1]]))
```

```python
svr_model, loaded_scaler = pickle.load(open("Insurance_premium_predictor_SVR.sav", "rb"))

# Scale new input before prediction
new_input = [[40,0,27.9,0,1]]
new_input_scaled = loaded_scaler.transform(new_input)

result = svr_model.predict(new_input_scaled)
print("Premium predicted by SVR model",result)
```

```python
tree_loaded_model = pickle.load(open("Insurance_premium_predictor_Decision_Tree.sav", 'rb'))
tree_result = tree_loaded_model.predict([[40,0,27.9,0,1]])
print("Premium predicted by Decision tree model",tree_result)
```

```python
forest_loaded_model = pickle.load(open("Insurance_premium_predictor_Random_Forest.sav", 'rb'))
forest_result = forest_loaded_model.predict([[40,0,27.9,0,1]])
print("Premium predicted by Random forest model", forest_result)
```
