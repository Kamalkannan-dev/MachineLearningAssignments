```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering

# Generate irregular dataset (two interleaving half circles)
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

# Spectral Clustering model
sc = SpectralClustering(
    n_clusters=2,          # Number of clusters you want
    affinity='nearest_neighbors',  # Similarity measure (can be 'rbf', 'nearest_neighbors', etc.)
    n_neighbors=10,        # Used when affinity='nearest_neighbors'
    assign_labels='kmeans',# Method to assign labels after spectral embedding
    random_state=42
)

labels = sc.fit_predict(X)

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.title("Spectral Clustering on Two Moons")
plt.show()
```
