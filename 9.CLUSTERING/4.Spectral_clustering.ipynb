```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering

# Generate irregular dataset (two interleaving half circles)
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

# Spectral Clustering model
sc = SpectralClustering(
    n_clusters=2,          # Number of clusters you want
    affinity='nearest_neighbors',  # Similarity measure (can be 'rbf', 'nearest_neighbors', etc.)
    n_neighbors=10,        # Used when affinity='nearest_neighbors'
    assign_labels='kmeans',# Method to assign labels after spectral embedding
    random_state=42
)

labels = sc.fit_predict(X)

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.title("Spectral Clustering on Two Moons")
plt.show()
```

```python
import numpy as np
from sklearn.datasets import make_circles

# --- Concentric Rings Dataset ---
X_rings, y_rings = make_circles(n_samples=400, factor=0.5, noise=0.05, random_state=42)

sc_rings = SpectralClustering(
    n_clusters=2,              # Expect 2 rings
    affinity='nearest_neighbors', # Graph-based similarity
    n_neighbors=10,
    assign_labels='kmeans',
    random_state=42
)
labels_rings = sc_rings.fit_predict(X_rings)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(X_rings[:,0], X_rings[:,1], c=labels_rings, cmap='viridis', s=30)
plt.title("Spectral Clustering on Concentric Rings")

# --- Spiral Dataset (synthetic) ---
theta = np.linspace(0, 4*np.pi, 400)
r = np.linspace(0.1, 1, 400)
X_spiral = np.vstack((r*np.cos(theta), r*np.sin(theta))).T
X_spiral += 0.05 * np.random.randn(*X_spiral.shape)  # add noise

sc_spiral = SpectralClustering(
    n_clusters=2,              # Expect 2 spiral arms
    affinity='nearest_neighbors',
    n_neighbors=10,
    assign_labels='kmeans',
    random_state=42
)
labels_spiral = sc_spiral.fit_predict(X_spiral)

plt.subplot(1,2,2)
plt.scatter(X_spiral[:,0], X_spiral[:,1], c=labels_spiral, cmap='plasma', s=30)
plt.title("Spectral Clustering on Spiral Data")

plt.show()
```
